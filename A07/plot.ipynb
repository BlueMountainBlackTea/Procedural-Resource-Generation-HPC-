```python
import matplotlib.pyplot as plt

# Data from the benchmarking table
threads = [1, 2, 4, 6, 8, 10, 12, 16, 20, 32, 64]
speedup = [0.9610, 1.9837, 2.4090, 2.7633, 2.9357, 3.2227, 2.5442, 2.9273, 3.0833, 3.5899, 3.7336]

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(threads, speedup, marker='o', linestyle='-', color='b')

# Add titles and labels
plt.title('Speedup vs. Number of Threads\n(DLA Simulation with 90,000 walkers on a 300x300 grid)')
plt.xlabel('Number of Threads')
plt.ylabel('Speedup')
plt.grid(True)

# Set x-ticks to match the threads values
plt.xticks(threads, fontsize = 8)

# Save the plot as a PNG file
plt.tight_layout()
plt.savefig('speedup_plot.png', dpi=300)

# Display the plot
plt.show()
```
